```python
import pandas as pd
import datetime

def date_parser_function(dateAsString):
    return datetime.datetime.strptime(dateAsString,"%m/%d/%Y")

df = pd.read_csv("~/Programming/Python/BT/tests/TimeSeries",
                index_col="Date",
                parse_dates=["Date"],
                date_parser=date_parser_function,
                sep="\t")
```

```python
from ipywidgets import interact
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
x = np.linspace(0, 1, 100)
y2 = df['H2']
y3 = df['H3']
y4 = df['H4']
```

```python
p = figure(title="my own interactive graph", plot_height=300, plot_width=600, y_range=(-5,5))
r2 = p.line(x, y2, color="#1f77b4", line_width=3, visible=True)
r3 = p.line(x, y3, color="#ff7f0e", line_width=3, visible=False)
r4 = p.line(x, y4, color="#2ca02c", line_width=3, visible=False)
```

```python
def update(**kwargs):
    for k,v in kwargs.items():
        if k == "2":
            r2.visible = v
            r2.legend = 'H2'
        if k == "3":
            r3.visible = v
            r3.legend = 'H3'
        if k == "4":
            r4.visible = v
            r4.legend = 'H4'
    push_notebook()
```

```python
show(p, notebook_handle=True)
```

```python
interact(update, **{str(x+2) : False for x in range(3)})
```
